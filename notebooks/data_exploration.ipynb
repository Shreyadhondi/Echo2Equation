<h1> Dwonloading the MathSpeech Dataset </h1>

```python
from datasets import load_dataset

# Load the MathSpeech dataset (train split)
dataset = load_dataset("AAAI2025/MathSpeech", split="train")

# Check number of samples
print(f"Total samples: {len(dataset)}")
```

<h2> Now let us visualize a few samples </h2>

```python
# Convert to DataFrame (avoids triggering audio decoding)
dataset_df = dataset.to_pandas()

# Show the first 5 rows
dataset_df.head(5)
```

<h2> Trimming the audip and the source columns. </h2>
<h5> We do not need the audio and the sourse columns to train out model for our project, so i am just trimming it off. </h5>

```python
import json

# Step 1: Select only the columns we need
trimmed_df = dataset_df[["transcription", "LaTeX"]]

# Step 2: Rename columns for consistency
trimmed_df = trimmed_df.rename(columns={"transcription": "spoken", "LaTeX": "latex"})

# Step 3: Save to JSONL file
output_path = "mathspeech_trimmed.jsonl"

with open(output_path, "w", encoding="utf-8") as f:
    for _, row in trimmed_df.iterrows():
        json.dump({"spoken": row["spoken"], "latex": row["latex"]}, f)
        f.write("\n")

print(f"Trimmed dataset saved to {output_path}")
```
